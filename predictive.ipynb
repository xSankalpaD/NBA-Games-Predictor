```python
import pandas as pd
```

```python
df= pd.read_csv("nba_games.csv", index_col=0) #index col lets the first row be index
```

```python
df= df.sort_values("date") #sorts based on date
```

```python
df= df.reset_index(drop=True) #rerste index adn drop removes old index
```

```python
del df["mp.1"]
del df["mp_opp.1"]
del df["index_opp"]
```

```python
def add_target(team):
    team["target"]= team["won"].shift(-1)
    return team

df= df.groupby("team", group_keys=False).apply(add_target)
```

```python
df["target"][pd.isnull(df["target"])] =2 #takes all the null value in target and makes it 2
```

```python
df["target"] = df["target"].astype(int, errors="ignore") #astype converts datatype in pandas
```

```python
#getting rid of nulls
nulls= pd.isnull(df)
```

```python
nulls= nulls.sum()
```

```python
nulls= nulls[nulls>0]
```

```python
valid_columns = df.columns[~df.columns.isin(nulls.index)]
```

```python
df= df[valid_columns].copy()
```

```python
#using feature selecter to selct small set of  columns to train the model
from sklearn.model_selection import TimeSeriesSplit
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import RidgeClassifier

rr= RidgeClassifier(alpha=1)
tsplit= TimeSeriesSplit(n_splits=3)

sfs= SequentialFeatureSelector(rr, n_features_to_select= 30, direction="forward", cv= tsplit)
```

```python
#scaling the columns to 0 or 1 (ridge regression works best scaled)
removed_columns = ["season", "date", "won", "target", "team", "team_opp"] # columns that doesnt make sense to scale
selected_columns = df.columns[~df.columns.isin(removed_columns)] #columns to scale
```

```python
#scaling the data (ridgeclassifier works best woth sclaed data)
from sklearn.preprocessing import MinMaxScaler

scaler= MinMaxScaler()
df[selected_columns] = scaler.fit_transform(df[selected_columns])
```

```python
df
```

```python
sfs.fit(df[selected_columns], df["target"])
```

```python
predictors = list(selected_columns[sfs.get_support()])
```

```python
predictors
```

```python
#this fucntion makes the predictions
def backtest(data, model, predictors, start=2, step=1):
    all_predictions = []
    
    seasons = sorted(data["season"].unique())
    
    for i in range(start, len(seasons), step):
        season = seasons[i]
        train = data[data["season"] < season]
        test = data[data["season"] == season]
        
        model.fit(train[predictors], train["target"])
        
        preds = model.predict(test[predictors]) #.predict return numpyarray so 
        preds = pd.Series(preds, index=test.index) #chnage to series
        combined = pd.concat([test["target"], preds], axis=1)
        combined.columns = ["actual", "prediction"]
        
        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
predictions=  backtest(df,rr,predictors)
```

```python
predictions
```
